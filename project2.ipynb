```python
import cv2
import matplotlib.pyplot as plt

# Assuming img, ClassIndex, bbox, classLabels are defined

font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1  # Adjust for size
thickness = 2
color = (0, 255, 0)  # Green text

for ClassInd, boxes in zip(ClassIndex.flatten(), bbox):
    boxes = tuple(map(int, boxes))  # Ensure integer coordinates
    cv2.rectangle(img, boxes, (255, 0, 0), 2)

    if ClassInd == 0:  # Skip background
        continue

    label = classLabels[ClassInd - 1]
    cv2.putText(img, label, (boxes[0] + 10, boxes[1] + 30), font, font_scale, color, thickness)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2

# Initialize video capture
cap = cv2.VideoCapture('vedio.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Try webcam if video file fails
    if not cap.isOpened():
        raise IOError('Cannot open video source')

# Font settings
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Break if video ends or error occurs

    # Object detection (replace 'model' with your actual model)
    # Assuming model.detect returns ClassIndex, confidence, bbox
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    print(ClassIndex)

    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            boxes = tuple(map(int, boxes))  # Ensure boxes are integers
            cv2.rectangle(frame, boxes, (255, 0, 0), 2)  # Draw rectangle

            if ClassInd != 0:  # Skip background class (if applicable)
                label = classLabels[ClassInd - 1] # assuming classLabels are defined elsewhere
                cv2.putText(frame, label, (boxes[0] + 10, boxes[1] + 40), font, font_scale, (0, 255, 0), 3)

    cv2.imshow("object detection by simplilearn", frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
